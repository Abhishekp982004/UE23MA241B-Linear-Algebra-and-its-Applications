```python
import imageio.v2 as imageio
import numpy as np
import matplotlib.pyplot as plt

# Load the image 
image = imageio.imread("/Users/abhishekp/Downloads/4454.jpg.webp")

# 2. Check the size of the image
print("Image shape:", image.shape)

# 3. Check the type of the array
print("Data type of the image array:", image.dtype)

# 4. Convert to Grayscale if the image is RGB
if len(image.shape) == 3:
    grayscale_image = np.mean(image, axis=2).astype(np.uint8)
else:
    grayscale_image = image.copy()

# 5. Lightening and Darkening the image
lightened = np.clip(grayscale_image + 50, 0, 255)  # Increase brightness
darkened = np.clip(grayscale_image - 50, 0, 255)  # Decrease brightness

# 6. Adjust Contrast
contrast_adjusted = np.clip(1.5 * grayscale_image - 50, 0, 255).astype(np.uint8)

# 7. Crop the image (center crop example)
h, w = grayscale_image.shape
cropped = grayscale_image[h//4:3*h//4, w//4:3*w//4]

# 8. Apply Matrix Transformations
flipped = np.flip(grayscale_image, axis=1)  # Horizontal flip
rotated = np.rot90(grayscale_image)  # Rotate 90 degrees

# 9. Display the modified images
fig, axes = plt.subplots(2, 4, figsize=(15, 8))
axes[0, 0].imshow(grayscale_image, cmap='gray')
axes[0, 0].set_title("Original Grayscale")
axes[0, 1].imshow(lightened, cmap='gray')
axes[0, 1].set_title("Lightened")
axes[0, 2].imshow(darkened, cmap='gray')
axes[0, 2].set_title("Darkened")
axes[0, 3].imshow(contrast_adjusted, cmap='gray')
axes[0, 3].set_title("Contrast Adjusted")
axes[1, 0].imshow(cropped, cmap='gray')
axes[1, 0].set_title("Cropped")
axes[1, 1].imshow(flipped, cmap='gray')
axes[1, 1].set_title("Flipped Horizontally")
axes[1, 2].imshow(rotated, cmap='gray')
axes[1, 2].set_title("Rotated 90")

for ax in axes.flat:
    ax.axis("off")

plt.show()
```
